```python
import pandas as pd

# Source
src_gizi_buruk = "https://raw.githubusercontent.com/FahrulID/balita-jawa-barat/main/dataset/Gizi%20Buruk/dinkes-od_17107_jumlah_balita_berdasarkan_kategori_balita_gizi_buruk_data.csv"
src_imunisasi_lengkap = "https://raw.githubusercontent.com/FahrulID/balita-jawa-barat/main/dataset/Imunisasi%20Dasar/dinkes-od_17464_jml_bayi_mendapat_imunisasi_dasar_lengkap__jk_data.csv"
src_dbd = "https://raw.githubusercontent.com/FahrulID/balita-jawa-barat/main/dataset/Kasus%20DBD/dinkes-od_18509_jml_kasus_penyakit_demam_berdarah_dengue_dbd__kabupate_data.csv"
src_diare = "https://raw.githubusercontent.com/FahrulID/balita-jawa-barat/main/dataset/Kasus%20Diare/dinkes-5-od_17577_jml_kasus_diare_pada_balita_dilayani__kabupatenkota_data.csv"
src_pneumonia = "https://raw.githubusercontent.com/FahrulID/balita-jawa-barat/main/dataset/Kasus%20Pneumonia/dinkes-5-od_17566_jml_penemuan_kasus_pneumonia_pada_balita__jk_data.csv"
src_perkiraan_pneumonia = "https://raw.githubusercontent.com/FahrulID/balita-jawa-barat/main/dataset/Perkiraan%20Kasus%20Pneumonia/dinkes-5-od_17565_jml_perkiraan_kasus_pneumonia_pada_balita__kabupatenko_data.csv"
src_stunting = "https://raw.githubusercontent.com/FahrulID/balita-jawa-barat/main/dataset/Stunting/dinkes-od_17147_jumlah_balita_stunting_berdasarkan_kabupatenkota_data.csv"
```

```python
# Load Dataset
df_gizi_buruk = pd.read_csv(src_gizi_buruk)
df_imunisasi_lengkap = pd.read_csv(src_imunisasi_lengkap)
df_dbd = pd.read_csv(src_dbd)
df_diare = pd.read_csv(src_diare)
df_pneumonia = pd.read_csv(src_pneumonia)
df_perkiraan_pneumonia = pd.read_csv(src_perkiraan_pneumonia)
df_stunting = pd.read_csv(src_stunting)

# Print Dataset
display(df_gizi_buruk)
display(df_imunisasi_lengkap)
display(df_dbd)
display(df_diare)
display(df_pneumonia)
display(df_perkiraan_pneumonia)
display(df_stunting)
```

```python
new_df_gizi_buruk = df_gizi_buruk.pivot(index=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], columns='kategori_gizi_buruk', values='jumlah_balita')
new_df_gizi_buruk.columns = new_df_gizi_buruk.columns.str.lower().str.replace(' ', '_')
display(new_df_gizi_buruk)

new_df_imunisasi_lengkap = df_imunisasi_lengkap.pivot(index=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], columns='jenis_kelamin', values='jumlah_bayi')
new_df_imunisasi_lengkap.columns = new_df_imunisasi_lengkap.columns.str.lower().str.replace(' ', '_')
new_df_imunisasi_lengkap['jumlah_bayi_imunisasi_lengkap'] = new_df_imunisasi_lengkap['laki-laki'] + new_df_imunisasi_lengkap['perempuan']
new_df_imunisasi_lengkap = new_df_imunisasi_lengkap.drop(columns=['laki-laki', 'perempuan'])
new_df_imunisasi_lengkap = new_df_imunisasi_lengkap.loc[(slice(None), slice(None), [2019, 2020, 2021]), :]
display(new_df_imunisasi_lengkap)

new_df_dbd = df_dbd[['kode_kabupaten_kota', 'nama_kabupaten_kota', 'jumlah_kasus', 'tahun']]
new_df_dbd = new_df_dbd.rename(columns={'jumlah_kasus': 'jumlah_kasus_dbd'})
new_df_dbd = new_df_dbd.set_index(['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'])
new_df_dbd = new_df_dbd.loc[(slice(None), slice(None), [2019, 2020, 2021]), :]
display(new_df_dbd)

new_df_diare = df_diare[['kode_kabupaten_kota', 'nama_kabupaten_kota', 'jumlah_kasus', 'tahun']]
new_df_diare = new_df_diare.rename(columns={'jumlah_kasus': 'jumlah_kasus_diare'})
new_df_diare = new_df_diare.set_index(['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'])
new_df_diare = new_df_diare.loc[(slice(None), slice(None), [2019, 2020, 2021]), :]
display(new_df_diare)

new_df_pneumonia = df_pneumonia.pivot(index=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], columns='jenis_kelamin', values='jumlah_kasus')
new_df_pneumonia.columns = new_df_pneumonia.columns.str.lower().str.replace(' ', '_')
new_df_pneumonia['jumlah_pneumonia'] = new_df_pneumonia['laki-laki'] + new_df_pneumonia['perempuan']
new_df_pneumonia = new_df_pneumonia.drop(columns=['laki-laki', 'perempuan'])
new_df_pneumonia = new_df_pneumonia.loc[(slice(None), slice(None), [2019, 2020, 2021]), :]
display(new_df_pneumonia)

new_df_perkiraan_pneumonia = df_perkiraan_pneumonia[['kode_kabupaten_kota', 'nama_kabupaten_kota', 'jumlah_kasus', 'tahun']]
new_df_perkiraan_pneumonia = new_df_perkiraan_pneumonia.rename(columns={'jumlah_kasus': 'jumlah_perkiraan_pneumonia'})
new_df_perkiraan_pneumonia = new_df_perkiraan_pneumonia.set_index(['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'])
new_df_perkiraan_pneumonia = new_df_perkiraan_pneumonia.loc[(slice(None), slice(None), [2019, 2020, 2021]), :]
display(new_df_perkiraan_pneumonia)

new_df_df_stunting = df_stunting[['kode_kabupaten_kota', 'nama_kabupaten_kota', 'jumlah_balita_stunting', 'tahun']]
new_df_df_stunting = new_df_df_stunting.set_index(['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'])
new_df_df_stunting = new_df_df_stunting.loc[(slice(None), slice(None), [2019, 2020, 2021]), :]
display(new_df_df_stunting)
```

```python
# check for nulls
print(new_df_gizi_buruk.isnull().sum())
print(new_df_imunisasi_lengkap.isnull().sum())
print(new_df_dbd.isnull().sum())
print(new_df_diare.isnull().sum())
print(new_df_pneumonia.isnull().sum())
print(new_df_perkiraan_pneumonia.isnull().sum())
print(new_df_df_stunting.isnull().sum())
```

```python
# merge all dataset
df = pd.merge(new_df_gizi_buruk, new_df_imunisasi_lengkap, on=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], how='outer')
df = pd.merge(df, new_df_dbd, on=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], how='outer')
df = pd.merge(df, new_df_diare, on=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], how='outer')
df = pd.merge(df, new_df_pneumonia, on=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], how='outer')
df = pd.merge(df, new_df_perkiraan_pneumonia, on=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], how='outer')
df = pd.merge(df, new_df_df_stunting, on=['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], how='outer')
df = df.reset_index()
df = df.sort_values(['kode_kabupaten_kota', 'tahun'])
display(df)

df.to_csv('data/dataset.csv', index=False)
```
